# Data Cleaning and Preprocessing

```python
# check for unique columns

df.nunique()
```

```python
df.shape
```

```python
# check the null values
df.isnull().sum()
```

```python
# Drop the unnecessary Column 

drop_cols = [
    "id", "title", "body", "currency", "fee", "has_photo",
    "price_display", "price_type", "address", "source", "time"
]
df = df.drop(columns=drop_cols, errors="ignore")
```

```python
# null values removes from target column

df.dropna(subset=["price"], inplace=True)
```

```python
# Fill missing numeric columns with median, categorical with mode

for col in df.columns:
    if df[col].dtype in ["float64", "int64"]:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else "Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
# check and remove the duplicate values

print("There are" ,df.duplicated().sum(), "values in dataset")

df.drop_duplicates(inplace=True)

print("After removing duplicate :", df.duplicated().sum())
```

```python
# Final check
print("Cleaned dataset shape:", df.shape)
print("\nRemaining columns:", list(df.columns))
print("\nColumn datatypes:")
print(df.dtypes)
print("\nMissing values after cleaning:")
print(df.isna().sum())
print("\n Duplicate values:")
print(df.duplicated().sum())
```

#### Ridge Regression

```python
from sklearn.linear_model import Ridge

kfold = KFold(n_splits=5, shuffle=True, random_state=2025)

pipeline = make_pipeline(StandardScaler(), Ridge(alpha=1.0))

r2_scores = cross_val_score(pipeline, x, y, cv=kfold, scoring='r2')
mse_scores = -cross_val_score(pipeline, x, y, cv=kfold, scoring='neg_mean_squared_error')

print("Ridge Regression - R²:", r2_scores.mean())
print("Ridge Regression - Std deviation:", r2_scores.std())
print("Ridge Regression - MSE:", mse_scores.mean())
```

#### ElasticeNet Regression

```python
from sklearn.linear_model import ElasticNet

pipeline = make_pipeline(StandardScaler(), ElasticNet(alpha=0.01)) 

kfold = KFold(n_splits=5, shuffle=True, random_state=2025)

r2_scores = cross_val_score(pipeline, x, y, cv=kfold, scoring='r2')
mse_scores = -cross_val_score(pipeline, x, y, cv=kfold, scoring='neg_mean_squared_error')

print("ElasticNet Regression - R²:", r2_scores.mean())
print("ElasticNet Regression - MSE:", mse_scores.mean())
```

#### KNN Regression

```python
from sklearn.neighbors import KNeighborsRegressor

pipeline = make_pipeline(StandardScaler(), KNeighborsRegressor(n_neighbors=5))  

kfold = KFold(n_splits=5, shuffle=True, random_state=2025)

r2_scores = cross_val_score(pipeline, x, y, cv=kfold, scoring='r2')
mse_scores = -cross_val_score(pipeline, x, y, cv=kfold, scoring='neg_mean_squared_error')

print("KNN Regression - R²:", r2_scores.mean())
print("KNN Regression - MSE:", mse_scores.mean())
```

#### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

# Random Forest does not need scaling
pipeline = make_pipeline(RandomForestRegressor(n_estimators=50, random_state=2025))

kfold = KFold(n_splits=5, shuffle=True, random_state=2025)

r2_scores = cross_val_score(pipeline, x, y, cv=kfold, scoring='r2')
mse_scores = -cross_val_score(pipeline, x, y, cv=kfold, scoring='neg_mean_squared_error')

print("Random Forest Regression - R²:", r2_scores.mean())
print("Random Forest Regression - MSE:", mse_scores.mean())
```

#### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

# Decision Tree does not need scaling
pipeline = make_pipeline(DecisionTreeRegressor(random_state=2025))

kfold = KFold(n_splits=5, shuffle=True, random_state=2025)

r2_scores = cross_val_score(pipeline, x, y, cv=kfold, scoring='r2')
mse_scores = -cross_val_score(pipeline, x, y, cv=kfold, scoring='neg_mean_squared_error')

print("Decision Tree Regression - R²:", r2_scores.mean())
print("Decision Tree Regression - MSE:", mse_scores.mean())
```

Observation:

Lower MSE is better. Random Forest again dominates.

Linear models have higher MSE than ElasticNet, Lasso, and Ridge, suggesting regularization helps a bit in high-dimensional data.

KNN has the highest MSE — consistent with poor R².

# Tuning the best Model 

#### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score
import numpy as np

# Define KFold strategy
kfold = KFold(n_splits=5, shuffle=True, random_state=2025)

# List of n_estimators to try
n_estimators_list = [50, 100, 200, 300, 400, 500]

# Store results
results = []

for n in n_estimators_list:
    rf = RandomForestRegressor(n_estimators=n, random_state=2025, n_jobs=-1)
    r2_scores = cross_val_score(rf, x, y, cv=kfold, scoring='r2')
    mse_scores = -cross_val_score(rf, x, y, cv=kfold, scoring='neg_mean_squared_error')
    
    results.append({
        "n_estimators": n,
        "R² Mean": r2_scores.mean(),
        "R² Std": r2_scores.std(),
        "MSE Mean": mse_scores.mean()
    })

# Convert to DataFrame for easy comparison
import pandas as pd
tuning_df = pd.DataFrame(results)
tuning_df = tuning_df.sort_values(by="R² Mean", ascending=False)
print(tuning_df)
```

### Key Takeways

Tree-based ensemble (Random Forest) is your best option here.

Regularized linear models (ElasticNet, Lasso, Ridge) are better than plain Linear Regression for this high-cardinality, high-dimensional dataset.

KNN is not suitable for datasets with thousands of one-hot encoded features.

Decision Tree alone is less robust than Random Forest because it overfits a single tree.
